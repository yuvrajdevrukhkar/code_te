```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data = [
    ['laptop','Eraser','Mouse','Scale'],
    ['Box','Colours','Mouse','Scale'],
    ['laptop','Keyboard','Eraser','Mouse']
]
```

```python
te = TransactionEncoder()
```

```python
transform_data = te.fit(data).transform(data)
```

```python
df = pd.DataFrame(transform_data,columns = te.columns_)
```

```python
df
```

```python
#build the model
frq_items = apriori(df, min_support = 0.6 , use_colnames = True)

#Collecting inferred rules in dataframe
rules = association_rules(frq_items ,metric="lift", min_threshold = 1)
```
